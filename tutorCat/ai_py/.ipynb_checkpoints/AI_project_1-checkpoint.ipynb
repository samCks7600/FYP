```python
file = 'steam-200k-cleaned.csv'
col = ['playerID', 'gameName', 'Action', 'playHours']
df = pd.read_csv(file, header = None,names = col )
df
```

```python
prototypeDf = df[(df['Action'] =="play")]
```

```python
prototypeDf
```

```python
prototypeDf = prototypeDf[prototypeDf['playHours'] >= 0.1]
```

```python
prototypeDf
```

```python
prototypeDf = prototypeDf[prototypeDf.groupby('gameName').playerID.transform(len)>=0]
prototypeDf
```

```python
print("Games Number : " + str(prototypeDf['gameName'].nunique()) )
```

```python
prototypeDf
```

playHoursAvgDf : col( gameName	playHours)

```python
# make Avg Table 
playHoursAvgDf = prototypeDf.groupby(['gameName'],as_index = False).playHours.mean()
```

```python
# playHoursAvgDf['playHours'] is avg now , then new a table playHoursAvgDf
playHoursAvgDf['avg_playHours'] = playHoursAvgDf['playHours']
playHoursAvgDf.drop(columns = 'playHours',inplace = True)
playHoursAvgDf.head()
```

```python
# let prototypeDf + playHoursAvgDf = FullColDf
FullColDf = prototypeDf.merge(playHoursAvgDf,on='gameName')
FullColDf
```

```python
RatingCondition = [
    FullColDf['playHours'] >= (1.25*FullColDf['avg_playHours']),# 5
    (FullColDf['playHours'] < (1.25*FullColDf['avg_playHours']))&(FullColDf['playHours'] >= (FullColDf['avg_playHours'])),# 4
    (FullColDf['playHours'] < (FullColDf['avg_playHours']))&(FullColDf['playHours'] >= (0.5*FullColDf['avg_playHours'])),# 3
    (FullColDf['playHours'] < (0.5*FullColDf['avg_playHours']))&(FullColDf['playHours'] >= (0.25*FullColDf['avg_playHours'])),# 2
    (FullColDf['playHours'] >= 0),# 1
]

# 125% = 5
# 124.9% ~100% = 4
# 99.9%~50% = 3
# 50%~25% = 2
# have play rating 1

RatingMark = [5,4,3,2,1]

FullColDf['rating'] = np.select(RatingCondition,RatingMark)

FullColDf
    
```

```python
FullColDf.drop(columns = ['Action','playHours','avg_playHours'], inplace = True)
FullColDf
```

```python
piv_dataFrame=FullColDf.pivot_table(index='playerID', columns='gameName',values="rating").fillna(0)
```

```python
piv_dataFrame
```

```python
def standardize(row):
    new_row = (row - row.mean()) / (row.max() - row.min())
    return new_row

piv_dataFrame = piv_dataFrame.apply(standardize)
```

```python
piv_dataFrame
```

```python
game_similarity = cosine_similarity(piv_dataFrame.T)
print(len(game_similarity))
```

```python
game_similarity_df = pd.DataFrame(game_similarity,index=piv_dataFrame.columns,columns=piv_dataFrame.columns)

game_similarity_df
```

```python
def get_similarity_game(game_name,user_rating):

    similarity_mark = game_similarity_df[game_name]*user_rating
    similarity_mark = similarity_mark.sort_values(ascending=False)

    return similarity_mark

print(get_similarity_game("Warframe",5))
```

```python
game_lover = [("Zombies Monsters Robots",5),("Blockstorm",3),("Blocks That Matter",1)]

print(type(game_lover))
print(game_lover)

similarity_games_tb = pd.DataFrame()

for game,rating in game_lover:
    similarity_games_tb = similarity_games_tb.append(get_similarity_game(game,rating),ignore_index=True)


similarity_games_tb.head()

similarity_games_tb.sum().sort_values(ascending = False)
```
